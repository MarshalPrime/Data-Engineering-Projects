
## Data Mining: Homework 1 - Python Implementation

### Question 1:

Download the excel file ”UntidyData Badly-Structured-Sales-Dat.csv” from Teams and re- arrange this data into the correct four columns. There has been a mix of rows and columns everywhere. Also, watch out for Grand Totals and Sub Totals, you do not need those in tidy data.


```python
import pandas as pd
import numpy as np
```

```python
file_path ='UntidyData_Badly-Structured-Sales-Dat.csv'
df = pd.read_csv(file_path)
```

```python
df.head()
```

Assignment is to rearrage this dataset in four column.

## 1. Dropping Total columns


```python
# Drop the 'Total' columns
data = df.drop(columns=['Consumer Total', 'Corporate Total', 'Home Office Total'])

data.head()
```

## 2. Creating New Columns

```python
order_type = sorted(['Consumer', 'Corporate', 'Home Office'] * 4)
order_type.insert(0, 'Order ID')
column_names =[f'{x}|{y}' for x, y in zip(order_type, data.iloc[0].values)]
column_names
```

```python
data.columns = column_names
```

```python
data.head()
```

```python
# Drop two rows 
rows_to_drop = [0, 1]
data = data.drop(index=rows_to_drop)

```

```python
data.head()
```

```python
data1 = pd.melt(data, id_vars=column_names[0], var_name='OrderType', value_name='Value')
data1.head()
```

```python
data1[['OrderType', 'Class']] = data1['OrderType'].str.split('|', expand=True)

data1.head()
```

```python
data1.shape
data1.info()
```

### Drop Null Values

```python
data2 = data1.dropna()
data2.head()
```

```python
data2.shape
data2.info()
```

## Rename columns

```python
Cleaned_data = data2.rename(columns={'Order ID|Ship Mode>>': 'OrderID', 'OrderType': 'OrderType', 'Value': 'Value','Class':'Class'})
```

```python
Cleaned_data.head()
```

## Rearrange Columns

```python
Cleaned_data= Cleaned_data [['OrderID', 'OrderType', 'Class', 'Value']]
```
